### 4)

#### Pseudo-code avec remise:

initialise w (randomly);

initialise j = 0;

while j < n_iter * n:

    i = choose randomly from [0 to n]
    
    w <- w - epsilon *  grad(loss(f(x(i)), y(i)))
    
    j <- j+1

#### Pseudo-code sans remise:

initialise w (randomly);

initialise j = 0;

while j < n_iter:

    shuffled_idx = shuffle [0, ... , n]
    
    for i in shuffled_idx:

        i = choose randomly from [0 to n]
    
        w <- w - epsilon * grad(loss(f(x(i)), y(i)))
    
    j <- j+1

#### Batch gradient descent with MSE as loss function

```python
w_ini = np.random.rand(3)-0.5
wh, loss_hist = gradient(X_clown, Y_clown, niter=2000, eps=0.0001, w_ini=w_ini,
                   loss_fun=mse_loss, gr_loss_fun=gr_mse_loss, stochastic=False)
```

```python
plt.figure(figsize=(14,10))
plt.title('Batch gradient descent for MSE')
plot_gradient(X_clown, Y_clown, wh, loss_hist, mse_loss)
```

For the stochastic case, for each iterations we use only one point of the observation sample, while for the batch gradient we use the whole data set (of size n), therefore in order to work with the comparable number of iterations we should multiple n_iter by n in for the stochastic case
